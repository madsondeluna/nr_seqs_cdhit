```python
import subprocess

# Defina o caminho para o arquivo FASTA que você subiu
input_fasta = 'original_seqs.fasta'
output_fasta = 'nr_seqs.fasta'  # sequências não redundantes

# Rodar CD-HIT com identidade de 90%
identity_threshold = 0.9

subprocess.run([
    'cd-hit',
    '-i', input_fasta,
    '-o', output_fasta,
    '-c', str(identity_threshold),
    '-n', '5'  # word-size 5 adequado para threshold 0.9 (veja abaixo explicação)
])
```

```python
# Função para analisar arquivo .clstr
def analyze_clusters(clstr_file):
    clusters = {}
    current_cluster = None

    with open(clstr_file, 'r') as f:
        for line in f:
            line = line.strip()
            if line.startswith('>Cluster'):
                current_cluster = line.split()[1]
                clusters[current_cluster] = []
            else:
                seq_name = line.split('>')[1].split('...')[0]
                clusters[current_cluster].append(seq_name)

    total_original = sum(len(v) for v in clusters.values())
    total_clusters = len(clusters)
    total_removed = total_original - total_clusters

    print(f"Total de sequências originais: {total_original}")
    print(f"Total de sequências após redução: {total_clusters}")
    print(f"Total de sequências removidas (redundantes): {total_removed}\n")

    print("Detalhes dos clusters e sequências agrupadas:\n")
    for cluster_id, seqs in clusters.items():
        if len(seqs) > 1:
            print(f"Cluster {cluster_id}:")
            print(f"  Representante: {seqs[0]}")
            print(f"  Sequências removidas:")
            for redundant_seq in seqs[1:]:
                print(f"    - {redundant_seq}")
            print()

# Arquivo clstr gerado pelo CD-HIT
clstr_file = 'nr_seqs.fasta.clstr'

# Executa a análise
analyze_clusters(clstr_file)
```
